Given new location data - calculate patent level distance betweeen inventors

```python
patent_distance_dict = dict.fromkeys(document_data["patent_number"])
for pat_num, pat_df in tqdm(document_data.groupby("patent_number")):

    if (pat_df["latitude"].isnull().values.any()) | (pat_df.iloc[0]["team_size"] == 1):

        patent_distance_dict[pat_num] = np.nan

    else:

        idxs = pat_df.index
        inventor_combos = list(combinations(idxs, 2))
        
        pat_distances = []
        for (inv1, inv2) in inventor_combos:

            d = distance.distance(tuple(pat_df.loc[inv1][["latitude", "longitude"]]), 
                                  tuple(pat_df.loc[inv2][["latitude", "longitude"]])).km
            pat_distances.append(d)

        patent_distance_dict[pat_num] = np.mean(pat_distances)
        
document_data["geo_distance"] = document_data["patent_number"].map(patent_distance_dict)
```
